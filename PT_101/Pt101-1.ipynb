```python
import pandas as pd 

df = pd.read_csv('pokemon_data.csv')

print(df.head(10))
```

```python
## Reading Headers
print('Printing The column headers')
print('-----------------------------------------------')
print('\n')
print(df.columns)
print('\n')
# Read a specific column
print("Printing the Name column")
print('-----------------------------------------------')
print(df['Name'])
print('\n')
```

```python
print('Print multiple selected columns')
print('-----------------------------------------------')
print(df[['Name', 'Type 1', 'HP']])
```

```python
# Print each row 
print('Printing each row')
#use iloc function which stands for integer location
print('-----------------------------------------------')
print(df.iloc[1:4]) # prints whatever data was present in the first row
```

```python
# read a specific location (R,C)
print('Printing a specific location')
print('-----------------------------------------------')
print(df.iloc[2,1])
```

```python
for index, row in df.iterrows():
    print(index,row['Name'])
```

```python
# Finding any sort of data
df.loc[df['Type 1'] == 'Fire']
```
